```python
import pandas as pd
import os
from itertools import chain


isnull = lambda x : True if len(str(x).strip())==0 \
                or (str(x).lower() in ['nan', 'none', r'\N', r'\n'])   \
                else False
                
notnull = lambda x : False if isnull(x) else True
extend_list = lambda x : list(chain(*x))

def all_alphabet(ss):
    for s in ss:
        if not((s >= u'\u0041' and s<=u'\u005a') or (s >= u'\u0061' and s<=u'\u007a')) :
                return False
    return True

def have_alphabet(ss):
    for s in ss:
        if (s >= u'\u0041' and s<=u'\u005a') or (s >= u'\u0061' and s<=u'\u007a'):
                return True
    return False 

def filter_text(s):
    if len(s.strip())<5:
        return False
    if not have_alphabet(s.strip()):
        return False
    return True

def word_filter(s):
    if len(s)<2:
        return False
    if not all_alphabet(s.replace(' ', '')):
        return False

    return True

def read_txt(path, line_split=True):
    try:
        with open(path, encoding = 'utf-8') as f:
            l = [x.strip() for x in f.readlines() if len(x.strip())!=0]
    except:
        with open(path, encoding='gbk') as f:
            l = [x.strip() for x in f.readlines() if len(x.strip())!=0]
    
    if line_split:
        return l
    else:
        return [' '.join(l)]



OUTPUT_DIR = r"Results"
if not os.path.exists(OUTPUT_DIR):
    os.mkdir(OUTPUT_DIR)

stopwords = read_txt(r"MyStopwords.txt")
setword_mine = read_txt(r"MySetwords.txt")

from nltk import word_tokenize
from nltk.tokenize import MWETokenizer

words = [xx.strip().split() for xx in setword_mine]
tokenizer1 = MWETokenizer(words, separator = ' ')
tokenizer = lambda x: [xx for xx in tokenizer1.tokenize(word_tokenize(str(x).lower())) if len(xx.strip())>0 and word_filter(xx.strip()) and (xx.strip() not in stopwords)]


from nltk import pos_tag
 
pos = lambda x : pos_tag(word_tokenize(x))

from nltk import sent_tokenize

cut_sent = lambda x : sent_tokenize(x)
```

```python
import pandas as pd

# 读取Excel文件
file_path = "Hecker.xlsx"
df = pd.read_excel(file_path)

# 提取 "content" 列的所有内容，忽略空值
list_res = [x for x in df['content'] if pd.notnull(x)]

# 输出结果
# print(list_res)
```

```python
df = pd.DataFrame()
df['text'] = list_res

def remove_whitespace(text):
    # 删除空格
    import re
    text = text.replace("_x000d_", " ")
    text = re.sub(r"\s+", " ", text)
    
    return text


# 大写转小写
df['text'] = df.apply(lambda x : str(x['text']).strip().lower().replace("\r", " "), axis=1)

# 删除多余空格
df['text'] = df.apply(lambda x : remove_whitespace(x['text']), axis=1)

print(df.head(5))
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

list_res = []

for text in list(df['text']):
    doc = nlp(text)

    list_temp = []
    for entity in doc.ents:
        list_temp.append(entity.text + ' ' + entity.label_)
    list_res.append('\n'.join(list_temp))

df['extraction'] = list_res


df.to_excel(os.path.join(OUTPUT_DIR, 'Hecker.xlsx'), index=False)
```
