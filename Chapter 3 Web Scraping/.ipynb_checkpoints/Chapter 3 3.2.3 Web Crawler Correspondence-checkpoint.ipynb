```python
# Get the URL for Letters

import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# 设置 Chrome WebDriver 路径
chrome_driver_path = '/Users/bbqnj35/下载/chromedriver_mac64/chromedriver'

# 创建 ChromeOptions 对象，并设置选项
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('executable_path=' + chrome_driver_path)

# 创建 Chrome WebDriver 实例，传入 ChromeOptions 对象
driver = webdriver.Chrome(options=chrome_options)

# 打开网页
url = 'https://sova.si.edu/record/fsa.a.01/ref61?q=*'
driver.get(url)

# 等待页面加载
time.sleep(5)  # 可以根据实际情况调整等待时间

# 查找指定的 ul 元素
try:
    ul_element = driver.find_element(By.XPATH, '/html/body/main/div/div[1]/div/div[2]/ul/li[2]/ul/li[1]/ul') 
    
    if ul_element:
        li_elements = ul_element.find_elements(By.TAG_NAME, 'li')
        base_url = 'https://edan.si.edu/slideshow/viewer/?eadrefid=FSA.A.01_'
        urls = []
        for li in li_elements:
            li_id = li.get_attribute('id')
            if li_id:
                full_url = base_url + li_id
                urls.append(full_url)
                print(full_url)
        
        # 写入文件，排除最后一个元素
        if urls:
            with open('output_urls.txt', 'w') as file:
                for url in urls[:-1]:
                    file.write(url + '\n')
    else:
        print("ul element not found")
except Exception as e:
    print(f"Error: {e}")

# 关闭浏览器窗口
driver.quit()
```

```python
#Get the Name of Correspondents

import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# 设置 Chrome WebDriver 路径
chrome_driver_path = '/Users/bbqnj35/下载/chromedriver_mac64/chromedriver'

# 创建 ChromeOptions 对象，并设置选项
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('executable_path=' + chrome_driver_path)

# 创建 Chrome WebDriver 实例，传入 ChromeOptions 对象
driver = webdriver.Chrome(options=chrome_options)

# 打开网页
url = 'https://sova.si.edu/record/fsa.a.01/ref61?q=*'
driver.get(url)

# 等待页面加载
time.sleep(5)  # 可以根据实际情况调整等待时间

try:
    # 循环从 li[1] 到 li[500] 获取 span 元素中的内容
    with open('name.txt', 'w') as file_name:
        for i in range(1, 501):  # 从 1 到 500
            try:
                # 构建 XPath 获取每个 li 元素中的 span 元素内容
                xpath = f'/html/body/main/div/div[1]/div/div[2]/ul/li[2]/ul/li[1]/ul/li[{i}]/span/span[3]'
                span_element = driver.find_element(By.XPATH, xpath)
                
                if span_element:
                    name = span_element.text.strip()
                    file_name.write(name + '\n')
                    print(f"Name {i}: {name}")
                else:
                    print(f"Span element {i} not found")
            
            except Exception as e:
                print(f"Error in li[{i}]: {e}")

    print("Data saved successfully.")

except Exception as e:
    print(f"Error: {e}")

# 关闭浏览器窗口
driver.quit()
```

```python
# 计算 name.txt 文件的行数
with open('name.txt', 'r') as file_name:
    num_lines_name = sum(1 for line in file_name if line.strip())

# 计算 output_urls.txt 文件的行数
with open('output_urls.txt', 'r') as file_urls:
    num_lines_urls = sum(1 for line in file_urls if line.strip())

print(f"Number of lines in name.txt: {num_lines_name}")
print(f"Number of lines in output_urls.txt: {num_lines_urls}")
```

```python
# Merge the two Files

import openpyxl

def read_file_lines(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
    return [line.strip() for line in lines]

def write_to_excel(output_urls_file, name_file, output_excel):
    output_urls = read_file_lines(output_urls_file)
    names = read_file_lines(name_file)

    if len(output_urls) != len(names):
        print("Error: The number of URLs does not match the number of names.")
        return

    wb = openpyxl.Workbook()
    sheet = wb.active

    # 写入表头
    sheet.append(["URL", "Name"])

    # 写入数据
    for url, name in zip(output_urls, names):
        sheet.append([url, name])

    # 保存表格
    wb.save(output_excel)
    print(f"Data successfully written to {output_excel}")

if __name__ == '__main__':
    output_urls_file = 'output_urls.txt'
    name_file = 'name.txt'
    output_excel = 'web_crawler.xlsx'

    write_to_excel(output_urls_file, name_file, output_excel)
```

```python
import openpyxl
import requests
from bs4 import BeautifulSoup

sheetHeaders = ["page_id","content", "name"]

class DataSheet(object):
    """
    管理表格类
    """

    def __init__(self, filename):
        self.filename = filename

        # 判断文件是否存在，没有就创建
        if not os.path.exists(self.filename):
            wb = openpyxl.Workbook()
            wb.save(self.filename)
        self.wb = openpyxl.load_workbook(self.filename)
        self.sheet = self.wb.active
        # 如果是空表格，插入表头
        if self.sheet.max_row == 1:
            self.sheet.append(sheetHeaders)

    def add_row(self, data):
        if self.sheet:
            self.sheet.append(data)

    def save(self):
        if self.wb:
            self.wb.save(self.filename)


def get_name(filename, url):
    wb = openpyxl.load_workbook(filename)
    sheet = wb.active
    for row in range(2, sheet.max_row + 1):
        if sheet.cell(row=row, column=1).value == url:
            return sheet.cell(row=row, column=2).value  # Assuming Name column is in column B (index 2)
    return None


def get_letter(url):
    dataSheet = DataSheet("Correspondence.xlsx")
    try:
        response = requests.get(url)
        html = BeautifulSoup(response.text, "html.parser")
        page_num = html.find("div", id="images", class_="grid").get("data-num-pages")
        
        page_num = int(page_num)

        for span in html.find_all("span", class_="metadata d-hide"):
            page_id = span.get("data-uan")
            content = span.get("data-transcription")
            
            # 如果 content 或 page_id 为 None，则置为空字符串
            if content is None:
                content = ""
            if page_id is None:
                page_id = ""
            
            # 获取对应的 Name，并写入 test.xlsx 的 NAME 列
            name = get_name('web_crawler.xlsx', url)
            dataSheet.add_row([page_id, content, name])

        for i in range(2, page_num + 1):  # corrected range to include all pages
            new_url = url + f"&page={i}"
            response = requests.get(new_url)
            html = BeautifulSoup(response.text, "html.parser")
            for span in html.find_all("span", class_="metadata d-hide"):
                page_id = span.get("data-uan")
                content = span.get("data-transcription")
                
                # 如果 content 或 page_id 为 None，则置为空字符串
                if content is None:
                    content = ""
                if page_id is None:
                    page_id = ""
                
                # 获取对应的 Name，并写入 test.xlsx 的 NAME 列
                name = get_name('web_crawler.xlsx', url)
                dataSheet.add_row([page_id, content, name])

    except Exception as e:
        print(f"Error processing {url}: {str(e)}")
    finally:
        dataSheet.save()


if __name__ == '__main__':
    # 读取 web_crawler.xlsx 中的 URL 列内容
    wb_crawler = openpyxl.load_workbook('web_crawler.xlsx')
    sheet_crawler = wb_crawler.active

    urls = []
    for i in range(2, sheet_crawler.max_row + 1):
        url = sheet_crawler.cell(row=i, column=1).value  # Assuming URL column is in column A (index 1)
        if url:
            urls.append(url)

    # 遍历每个 URL 进行爬取并写入数据
    for url in urls:
        get_letter(url)
```
